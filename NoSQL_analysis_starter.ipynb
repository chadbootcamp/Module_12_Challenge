### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
# Query to find establishments in London with RatingValue >= 4
london_establishments = establishments.find({
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": 4}
})

# Convert the query result to a list and print the number of establishments found
london_establishments_list = list(london_establishments)
print(f"Number of establishments in London with RatingValue >= 4: {len(london_establishments_list)}")

# Optionally, display the first few results
for establishment in london_establishments_list[:5]:  # Print first 5 results
    pprint(establishment)

# Use count_documents to display the number of documents in the result
# Count the number of establishments in London with RatingValue >= 4
london_count = establishments.count_documents({
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": 4}
})

# Print the count of establishments found
print(f"Number of establishments in London with RatingValue >= 4: {london_count}")

# Display the first document in the results using pprint
from pprint import pprint

# Find the first establishment in London with RatingValue >= 4
first_london_establishment = establishments.find_one({
    "LocalAuthorityName": "London",
    "RatingValue": {"$gte": 4}
})

# Print the first document using pprint
if first_london_establishment:
    pprint(first_london_establishment)
else:
    print("No establishments found in London with RatingValue >= 4.")
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
from pprint import pprint

# Define the target latitude and longitude
target_latitude = 51.5074  # Example latitude for London
target_longitude = -0.1278  # Example longitude for London

# Define the range for latitude and longitude
latitude_range = {
    "$gte": target_latitude - 0.01,
    "$lte": target_latitude + 0.01
}

longitude_range = {
    "$gte": target_longitude - 0.01,
    "$lte": target_longitude + 0.01
}

# Query to find establishments within the specified range, with RatingValue = 5, and sort by hygiene score
nearby_establishments = establishments.find({
    "geocode.latitude": latitude_range,
    "geocode.longitude": longitude_range,
    "RatingValue": 5
}).sort("scores.Hygiene")  # Ensure the correct field name is used

# Convert the query result to a list
nearby_establishments_list = list(nearby_establishments)

# Print the number of establishments found
print(f"Number of establishments found within 0.01 degrees with RatingValue = 5: {len(nearby_establishments_list)}")

# Print the results
for establishment in nearby_establishments_list:  # Print all results
    pprint(establishment)
```

```python
# Convert result to Pandas DataFrame
import pandas as pd

# Convert the query result to a list
nearby_establishments_list = list(nearby_establishments)

# Convert the result to a Pandas DataFrame
nearby_establishments_df = pd.DataFrame(nearby_establishments_list)

# Display the first few rows of the DataFrame
print(nearby_establishments_df.head())
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
import pandas as pd

# Create the aggregation pipeline
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "count": {"$sum": 1}
        }
    },
    {
        "$sort": {
            "count": -1  # Sort from highest to lowest
        }
    }
]

# Execute the aggregation pipeline
results = establishments.aggregate(pipeline)

# Convert the results to a list
results_list = list(results)

# Convert the result to a Pandas DataFrame
results_df = pd.DataFrame(results_list)

# Rename the columns for better readability
results_df.rename(columns={"_id": "LocalAuthorityName"}, inplace=True)

# Display the number of rows in the DataFrame
print(f"Number of rows in the DataFrame: {results_df.shape[0]}")

# Display the first 10 rows of the DataFrame
print("First 10 rows of the DataFrame:")
print(results_df.head(10))
```
